# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
data = {
    "Year": list(range(2015, 2025)),
    "Value": [50, 52, 55, 60, 68, 80, 95, 110, 130, 150]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,4))

plt.plot(df["Year"], df["Value"], linewidth=2)
plt.scatter(df["Year"].iloc[-1], df["Value"].iloc[-1])
plt.text(2024, 150, "Peak growth", fontsize=10, ha='left')
plt.title("Sales Have Increased Rapidly Over Time", fontsize=14)

plt.xlabel("")
plt.ylabel("")
plt.grid(False)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.show()
```
